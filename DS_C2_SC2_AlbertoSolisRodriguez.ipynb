## Reto | Análisis del progreso mundial de vacunación

Se obtuvo información proveniente del archivo "country_vaccinations" obtenido de Kaggle y se construyó un script que permitió cumplir con los requerimientos indicados. Cada requerimiento tendrá su sección de texto y código individual.

*_El archivo se subió a la plataforma de Google Colab bajo la dirección "/content/sample_data/country_vaccinations.csv. Para el correcto funcionamiento del código debe utilizarse Google Colab como ambiente de ejecución."_

```python
# Antes de comenzar se deben importar NumPy y Pandas

import numpy as np
import pandas as pd
```

##### A. Extraer información del archivo.



```python
# Se asigna la información del archivo a un dataframe.

vaccinations = pd.read_csv("/content/sample_data/country_vaccinations.csv")
vaccinations
```

#### B. Mostrar la estructura y tipos de datos de cada columna para identificar qué operaciones puedes realizar con cada una de ellas, asegurándote que las columnas con fechas sean del tipo datetime64.

```python
# Se muestra la estructura de los datos

vaccinations.info()

# Las columnas con fechas no son del tipo datetime64 por lo que hay que corregirlo.

vaccinations["date"] = pd.to_datetime(vaccinations["date"])

# Se verifica que sea el tipo de dato correcto

vaccinations.info()
```

#### C. Determinar la cantidad de vacunas aplicadas de cada compañía (con base en cómo lo reporta cada país en la columna vaccines, en otras palabras, agrupe por vaccines y realice la sumatoria).

```python
# Se agrupa por las vacunas de cada compañía y se realiza la suma total de las vacunas puestas.

vaccinations[["vaccines","total_vaccinations"]].groupby("vaccines").sum()
```

#### D. Obtener la cantidad de vacunas aplicadas en todo el mundo.

```python
# Se eligió la cantidad diaria de vacunas para hacer la sumatoria para evitar el doble conteo que ocurriría si se utiliara "total_vaccinations".
cantidad = vaccinations["daily_vaccinations"].sum()

print(f"La cantidad de vacunas aplicadas en todo el mundo fue de {cantidad}")
```

#### E. Calcular el promedio de vacunas aplicadas por país.

```python
# Aquí se utiliza "total_vaccinations" para obtener el promedio de vacunas totales.
vaccinations[["country","total_vaccinations"]].groupby("country").mean()
```

#### F. Determinar la cantidad de vacunas aplicadas el día 29/01/21 en todo el mundo.

```python
# Se debe prestar atención a que el formato de fecha del dataframe no es el mismo al dado por los requerimientos.

cantidad_vacunas = vaccinations.loc[vaccinations["date"] == "2021-01-29", "daily_vaccinations"].sum()

print(f"La cantidad de vacunas aplicadas el 29-01-21 fue de {cantidad_vacunas}")
```

#### G. Crear un dataframe nuevo denominado "conDiferencias" que contenga los datos originales y una columna derivada (diferencias) con las diferencias de aplicación entre las columnas daily_vaccionations y daily_vaccionations_raw.

```python
# Primero se crea un nuevo dataframe sobre el que trabajar para evitar modificar el original. Posteriormente se realizan las operaciones
# sobre el nuevo dataframe (conDiferencias).

conDiferencias = vaccinations.copy()
conDiferencias["diferencias"] = conDiferencias["daily_vaccinations"] - conDiferencias["daily_vaccinations_raw"]

conDiferencias
```

#### H. Obtener el periodo de tiempo entre el registro con fecha más reciente y el registro con fecha más antigua.

```python
# Es suficiente con realizar una resta entre los registros.
periodo = vaccinations["date"].max() - vaccinations["date"].min()

print(f"El periodo de tiempo entre el registro más reciente y el más antiguo es de {periodo}.")
```

#### I.  Crear un dataframe nuevo denominado "conCantidad" que contenga los datos originales y una columna derivada (canVac) con la cantidad de vacunas utilizadas cada día (usar la columna vaccines y separar por el carácter , ).

```python
# Se realizó una copia del dataframe original para no afectarlo. Importante recordar el ".str" al tratar con métodos de strings.
conCantidad = vaccinations.copy()
conCantidad["canVac"] = conCantidad["vaccines"].str.split(", ").str.len()
conCantidad
```

#### J. Generar un dataframe denominado "antes20" con todos los registros que se hayan realizado antes del 20 de diciembre de 2020.

```python
# "query" facilita la construcción del nuevo dataframe. Debe prestarse atención al uso de comillas para que el programa no trate a la fecha dentro de "query"
# como un entero.
antes20 = vaccinations.copy()
antes20 = antes20.query("date < '2020-12-20'")
antes20
```

#### K. Obtener un dataframe denominado "pfizer" con todos los registros donde se haya utilizado la vacuna Pfizer.

```python
# Se obtuvieron los registros donde se hubiera utilizado la vacuna Pfizer, aunque fuera en combinación con otras.
pfizer = vaccinations.copy()

# Se creó una columna donde se menciona si se cumple con la condición de usar esa vacuna y se utiliza para filtrar los registros en los que no.
pfizer["contiene"] = pfizer["vaccines"].str.contains("pfizer", case = False)
pfizer = pfizer[pfizer["contiene"] == True]

# Ya que se tienen los registros que sí utilizaron la vacuna, ya no es necesaria la columna condicional.
del pfizer["contiene"]

pfizer
```

#### L. Almacenar los dataframes generados (conDiferencias, conCantidad, antes20 y pfizer) en un archivo de Excel denominado resultadosReto.xlsx, donde cada dataframe ocupe una hoja diferente.

```python
# Se usa ExcelWriter para poder "escribir" los dataframes en un archivo de Excel, cada uno en una hoja diferente.
# Cabe recalcar que la dirección "/content/sample_data/resultadosReto.xlsx" sólo funciona al ejecturar el código en
# un ambiente de Google Colab.

with pd.ExcelWriter("/content/sample_data/resultadosReto.xlsx") as writer:
  conDiferencias.to_excel(writer, sheet_name = "conDiferencias", index = False)
  conCantidad.to_excel(writer, sheet_name = "conCantidad", index = False)
  antes20.to_excel(writer, sheet_name = "antes20", index = False)
  pfizer.to_excel(writer, sheet_name = "pfizer", index = False)
```
